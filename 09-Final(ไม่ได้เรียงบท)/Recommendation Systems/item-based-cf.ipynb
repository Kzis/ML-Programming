## Item-Based Collaborative Filtering

```python
import numpy as np
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv', usecols=range(3))
movies = pd.read_csv('movies.csv', usecols=range(2))
```

```python
df = pd.merge(movies, ratings)
```

```python
df.head()
```

```python
movie_ratings = df.pivot_table(index=['userId'], columns=['title'], values='rating')
```

```python
movie_ratings.head()
```

```python
corr_mat = movie_ratings.corr()
corr_mat.head() 
```

```python
corr_mat = movie_ratings.corr(method='pearson', min_periods=100)
corr_mat.head() 
```

### Making recommendation for the user

```python
d = {'Forrest Gump (1994)': 5, 'Matrix, The (1999)': 5}
my_ratings = pd.Series(d)
my_ratings.index.name = 'title'
```

```python
my_ratings.index[1]
```

```python
corr_mat[my_ratings.index[0]].dropna().sort_values(ascending=False)
```

```python
sim_candidates = pd.Series()
for i in range(0, len(my_ratings.index)):
    print('Adding sims for', my_ratings.index[i], '...')
    sims = corr_mat[my_ratings.index[i]].dropna()
    sims = sims.map(lambda x: x * my_ratings[i])
    sim_candidates = sim_candidates.append(sims)
print('Sorting...')
sim_candidates.sort_values(inplace=True, ascending=False)
print(sim_candidates[:10])
```

```python
sim_candidates['Inception (2010)']
```

Some of the same movies showed up more than once, because they were similar to more than one movie the user rated. We'll use groupby() to add together the scores from movies that show up more than once, so they'll count more:

```python
sim_candidates = sim_candidates.groupby(sim_candidates.index).sum()
```

```python
sim_candidates.sort_values(inplace=True, ascending=False)
sim_candidates.head(10)
```

The last thing we have to do is filter out movies the user have already rated, as recommending a movie the user have already watched is not helpful:

```python
filtered_sims = sim_candidates.drop(my_ratings.index)
filtered_sims.head(10)
```
