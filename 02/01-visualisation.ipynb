## Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 50)
y = x**2
```

```python
fig = plt.figure()
```

```python
# left, bottom, width, height (range 0 to 1)
axes = fig.add_axes([0.1, 0.1, 0.9, 0.9])
```

```python
axes.plot(x, y, 'b')
axes.set_xlabel('X')
axes.set_ylabel('y')
axes.set_title(r'$y=x^2$')
```

```python
fig
```

```python
fig = plt.figure()
```

```python
axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes
```

```python
axes1.plot(x, y, 'b')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title(r'$y=x^2$')
```

```python
axes2.plot(y, x, 'r')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title(r'$y=\sqrt{x}$')
```

```python
fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes[0].plot(x, y, 'b')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_title(r'$y=x^2$')
```

```python
axes[1].plot(y, x, 'r')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
axes[1].set_title(r'$y=\sqrt{x}$')
```

```python
fig
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label=r"$x^2$")
ax.plot(x, x**3, label=r"$x^3$")
ax.legend()
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2015', periods=1000))
ts = ts.cumsum()
```

```python
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('ABCD'))
df = df.cumsum()
df.plot()
```
